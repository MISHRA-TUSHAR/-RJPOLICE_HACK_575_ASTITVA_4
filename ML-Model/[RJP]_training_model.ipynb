```python
from PyPDF2 import PdfReader
from langchain.embeddings import SentenceTransformerEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.chains.question_answering import load_qa_chain
from ctransformers import AutoModelForCausalLM
from transformers import AutoTokenizer, pipeline
from langchain.embeddings import HuggingFaceEmbeddings
import chromadb
from chromadb.config import Settings
from langchain.vectorstores import Chroma
```

```python
doc_reader= '/content/message.txt'
with open(doc_reader, 'r') as file:
    raw_text = file.read()
# Splitting up the text into smaller chunks for indexing
text_splitter = CharacterTextSplitter(
    separator = "\n",
    chunk_size = 100,
    chunk_overlap  = 20, #striding over the text
    length_function = len,
)
texts = text_splitter.split_text(raw_text)
```

```python
embeddings = SentenceTransformerEmbeddings(model_name="sentence-transformers/multi-qa-mpnet-base-dot-v1")
```

```python
query = 'Person A tried to murder Person B after stealing their Wealth and Raping them, what are the applicable acts and sectionas according to the IPC for person a to be Charged'
```

```python
docsearch = FAISS.from_texts(texts, embeddings)
docs_query = docsearch.similarity_search(query)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    "TheBloke/Mistral-7B-Instruct-v0.1-GGUF",
    model_file="mistral-7b-instruct-v0.1.Q4_K_M.gguf",
    model_type="mistral",
    gpu_layers=50,
    hf=True
)

# Tokenizer
tokenizer = AutoTokenizer.from_pretrained("mistralai/Mistral-7B-Instruct-v0.1")

# Pipeline
generator = pipeline(
    model=model, tokenizer=tokenizer,
    task='text-generation',
    max_new_tokens=512,
    repetition_penalty=1.1,
    do_sample = True,
    temperature = 0.3,
    top_p= 0.95
)
```
